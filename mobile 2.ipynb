```python
import numpy as np
from IPython.display import display
import pandas as pd
```

```python
pd.read_csv("https://raw.githubusercontent.com/pythonai250824/23.03.2025/main/mobile.csv")
```

```python
# תשובה לשאלה 1
# צור עמודה של שם פרטי מהמילה הראשונה של שם המנוי , השתמש בפקודת apply

def first_name(fn):
    return fn.split()[0]

df['First Name'] = df['Customer Name'].apply(first_name)
df
```

```python
# תשובה לשאלה 2
#050 פלפון, 051 גולן, 052 סלקום, 053 מוטורלה
# רמז: replace צור עמודה של חברת סלולר לפי תחילית מספר הטלפון:

df['Cellular Company'].replace({'050': 'Pelephone', '051': 'Golan', '052': 'Cellcom', '053': 'Motorola'})
df
```

```python
# תשובה לשאלה 3
#השתמש בעמודה שיצרת בסעיף הקודם וחשב:
# a. כמה מנויים יש לכל חברת סלולר . רמז: counts value
display('כמות מנויים לכל חברה:',df['Cellular Company'].value_counts())
# b. מהם השמות של החברות הסלולריות השונות? רמז: unique
display('שמות החברות השונות:',df['Cellular Company'].unique()) 
# c. כמה חברות סלולריות שונות יש? רמז: nunique 
display('כמה חברות שונות יש:',df['Cellular Company'].nunique()) 
```

```python
# תשובה לשאלה 5
# מיין את המנויים לפי זמן שיחה מהקטן לגדול
df[~df['Talked'].duplicated()].sort_values(by='Talked') 
```

```python
# תשובה לשאלה 5-המשך
# מיין את המנויים לפי זמן שיחה + זמן גלישה
df['Call_Browse'] = df['Talked'] & df['Internet']
df.sort_values(by='Call_Browse', ascending=True)
display(df)
```

```python
# תשובה לשאלה 6
# הדפס את 5 הדברנים הגדולים ביותר . רמז: nlargest
display('5 הדברנים:',df.nlargest(5,'Talked'))
# הדפס את 3 המנויים שגולשים הכי פחות. רמז: nsmallest
display('3 המנויים:',df.nsmallest(3,'Internet'))
```

```python
# תשובה לשאלה 7
# נראה שחלק מהרשומות נרשמו מספר פעמים-
# a. הדפס את הרשומות הכפולות
# הדפסת הרשומות הכפולות
display(df[df.duplicated()])

# b. הסר את הרשומות הכפולות
display(df['Total Time'].drop_duplicates()) 
display(df['Call + Browse'].drop_duplicates()) 
```
